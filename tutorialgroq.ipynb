```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.indeed.com/q-Web-Portal-jobs.html?vjk=98c48720a6924274")
docs = loader.load().pop().page_content
print(docs)
```

```python
from langchain_core.prompts import PromptTemplate
prompt_extract = PromptTemplate.from_template(
"""
### SCRAPED TEXT FROM WEBSITE:
{page_data}
### INSTRUCTION:
The scraped text from the career's page of a website.
Your job is to extract the job postings and return them in JSON format containing the
following keys: `role`, `experience`, `skills and description` .
Only return the valid JSON.
### VALID JSON (NO PREAMBLE):"""
)
```

```python
chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={'page_data':docs})
type(res.content)
```

```python
import pandas as pd
df = pd.read_csv("my_portfolio.csv")
df
```

```python
import chromadb
import uuid
# Initialize the ChromaDB client and connect to the 'vectorstore'
client = chromadb.PersistentClient('vectorstore')

# Fetch or create the 'portfolio' collection
collection = client.get_or_create_collection(name="portfolio")

# If the collection is empty, proceed to add documents
if not collection.count():  # Remove any unexpected indentation
    for index, row in df.iterrows():  # Properly indent this line
        # Add documents, metadata, and a unique ID for each entry
        collection.add(
            documents=[row["Techstack"]],  # Wrap 'Techstack' in a list
            metadatas={"links": row["Links"]},
            ids=[str(uuid.uuid4())]  # Use 'uuid4' for a random UUID
        )
```

```python
# Check the count of documents in the collection
document_count = collection.count()
print(f"Number of documents in the collection: {document_count}")

# Retrieve all documents to verify their content (if the collection is not too large)
if document_count > 0:
    all_documents = collection.query(query_texts=["*"], n_results=document_count)
    print("Documents in the collection:")
    for doc in all_documents.get('documents'):
        print(doc)
else:
    print("No documents found in the collection.")
```
